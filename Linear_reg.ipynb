```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

area = [2600, 3000, 3200, 3600, 4000]
price = [550000, 565000, 610000, 680000, 725000]
dataset = pd.DataFrame(area, columns=['area'])
dataset['price'] = price
dataset
```

```python
plt.scatter(dataset['area'], dataset['price'], color = 'red', marker = '+')
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.show()
```

```python
reg = LinearRegression()
reg.fit(dataset[['area']],dataset['price'])
```

```python
reg.predict([[3400]])
```

```python
reg.score(dataset[['area']], dataset['price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(dataset['area'], dataset['price'], color = 'red', marker = '+')
plt.xlabel('area(sq ft)')
plt.ylabel('price($)')
plt.plot(dataset['area'], reg.predict(dataset[['area']]), color = 'green')
plt.show()
```
