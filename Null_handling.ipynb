```python
import pandas as pd
df = pd.read_csv('sample.csv')
df.head()
```

```python
#remove any duplicate rows from the DataFrame
df = df.drop_duplicates()
df
```

```python
#Detecting Missing Values in a DataFrame
df.isnull().sum()
```

```python
#calculates the total number of missing values across the entire DataFrame
df.isnull().sum().sum()
```

```python
#Deleting the rows with null values
df1 = df.dropna() #Default axis=0
df1
```

```python
#deleting entire columns with null values
df2 = df.dropna(axis=1)
df2
```

```python
#Deleting a row which contains atleast 1 nul value
df3 = df.dropna(how = 'any')
df3
```

```python
#Deleting a row only if all values in rows are null
df4 = df.dropna(how = 'all')
df4
```

```python
print(df.shape)
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
```
