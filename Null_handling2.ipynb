```python
import pandas as pd
df = pd.read_csv('sample.csv')
df
```

```python
#Filling the null values
#df.fillna(0)
df.fillna(35)
```

```python
#It performs a forward fill in null values in axis 0 (As default)
df.fillna(method="ffill")
```

```python
#It performs forward fill in axis 1
df.fillna(method = "ffill", axis = 1)
```

```python
#It performs a backward fill in the null values
df.fillna(method = "bfill")
```

```python
physics_mean = df['Physics'].mean().round(2)
chemistry_mean = df['Chemistry'].mean().round(2)
maths_mean = df['Maths'].mean().round(2)
comp_mean = df['Computer'].mean().round(2)
print(physics_mean)
print(chemistry_mean)
print(maths_mean)
print(comp_mean)
```

```python
df['Physics'].fillna(value = physics_mean, inplace = True)
df['Chemistry'].fillna(value = chemistry_mean, inplace = True)
df['Maths'].fillna(value = maths_mean, inplace = True)
df['Computer'].fillna(value = comp_mean, inplace = True)
df
```
